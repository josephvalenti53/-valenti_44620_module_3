```python
import matplotlib.pyplot as plt
import numpy as np

# axes data
insertion_sort_times_y = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times_y = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]
sizes_x = [10, 20, 50, 100, 1000, 5000]

#plot creation
fig, ax1=plt.subplots()
ax2 = ax1.twinx()
ax1.plot(sizes_x, insertion_sort_times_y, color='red')
ax2.plot(sizes_x, merge_sort_times_y, color='blue')

ax1.set_xlabel('Sizes')
ax1.set_ylabel('Insertion Sort Times', color='red')
ax2.set_ylabel('Merge Sort Times', color='blue')

plt.show

```
